```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pizza_size =[[2],[3],[3.6],[4],[5],[2],[3.5]]
price= [[100],[120],[150],[175],[200],[120],[155]]
```

```python
from matplotlib.lines import MarkerStyle
plt.figure()
plt.title("PIZZA PRICE ACCORDING TO SIZE")
plt.xlabel('SIZE')
plt.ylabel('PRICE')
plt.plot(pizza_size,price,'bo',markeredgecolor='yellow')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(pizza_size,price)
```

```python
model.predict([[5]])
```
